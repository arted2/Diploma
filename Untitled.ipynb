```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2 = r2_score(y_true, y_pred)

    return {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'MAPE (%)': mape,
        'R2 Score': r2
    }
```

```python
import pandas as pd
import numpy as np


df = df.sort_values('Год').reset_index(drop=True)


df = df.set_index('Год')

# Создаём новый индекс с шагом 0.25 года 
full_index = np.arange(df.index.min(), df.index.max() + 0.25, 0.25)
df = df.reindex(full_index)

# Интерполируем все числовые значения
df_interpolated = df.interpolate(method='linear')


df_interpolated = df_interpolated.reset_index().rename(columns={'index': 'Год'})

df_interpolated['Год'] = df_interpolated['Год'].round(2)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Признаки и целевая переменная
X = df[['Ковид', 'Санкции', 'Ключевая ставка ЦБ РФ (%)', 'Доллар', 'ИПЦ (% к декабрю предыдущего года)']]
y = df['Квартиры']

# Словарь моделей
models = {
    'MLP': MLPRegressor(hidden_layer_sizes=(64, 32), activation='relu', max_iter=1000, random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=200, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=10000, learning_rate=0.05, max_depth=2, random_state=42)
}

# Сравнение моделей
results = []
predictions = {}

for name, model in models.items():
    model.fit(X, y)
    y_pred = model.predict(X)
    predictions[name] = y_pred

    mae = mean_absolute_error(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y - y_pred) / y)) * 100
    r2 = r2_score(y, y_pred)

    results.append({
        'Model': name,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'MAPE (%)': mape,
        'R2 Score': r2
    })

metrics_df = pd.DataFrame(results).sort_values(by='RMSE')
print("📊 Сравнение моделей по метрикам:\n", metrics_df)
```

```python
# Добавим предсказания в DataFrame
df_plot = df[['Год', 'Квартиры']].copy()
for name in predictions:
    df_plot[name] = predictions[name]

# Построим график
plt.figure(figsize=(12, 6))
plt.plot(df_plot['Год'], df_plot['Квартиры'], label='Фактические', linewidth=2, marker='o')

for name in predictions:
    plt.plot(df_plot['Год'], df_plot[name], label=f'{name}', linestyle='--', marker='s')

plt.title('Цены на квартиры: фактические и предсказания моделей')
plt.xlabel('Год')
plt.ylabel('₽/м²')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

# Параметры, которые будем перебирать
param_grid = {
    'n_estimators': [5,10,100, 200, 300, 1000, 10000],
    'max_depth': [2, 3, 4, 5, 15, 100],
    'learning_rate': [0.01, 0.05, 0.1, 0.2]
}

# Базовая модель
gbr = GradientBoostingRegressor(random_state=42)

# GridSearch с 5-кратной кросс-валидацией
grid_search = GridSearchCV(
    estimator=gbr,
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)

# Обучаем на тренировочной выборке
grid_search.fit(X_train, y_train)

# Лучшие параметры
best_params = grid_search.best_params_
print("🔍 Лучшие параметры:", best_params)

# Лучшая модель
best_gbr = grid_search.best_estimator_

# Предсказание на тесте
y_pred = best_gbr.predict(X_test)

# Оценка

evaluate_model(y_test, y_pred)
```

```python
from prophet import Prophet
import pandas as pd

# Подготовка данных
df_prophet = df_interpolated[['Год', 'Квартиры']].copy()
df_prophet = df_prophet.rename(columns={'Год': 'ds', 'Квартиры': 'y'})

# Преобразуем "год.доля" в реальную дату: например, 2024.75 -> октябрь 2024
def convert_decimal_year_to_datetime(decimal_year):
    year = int(decimal_year)
    remainder = decimal_year - year
    month = int(remainder * 12) + 1  # +1 чтобы январь был 1
    return pd.Timestamp(year=year, month=month, day=1)

# Применяем
df_prophet['ds'] = df_prophet['ds'].apply(convert_decimal_year_to_datetime)

# Инициализируем и обучаем модель Prophet
model = Prophet()
model.fit(df_prophet)

# Создаём будущие точки (ещё на 8 кварталов = 2 года)
future = model.make_future_dataframe(periods=8, freq='Q')

# Прогноз
forecast = model.predict(future)

# Визуализация
model.plot(forecast)
```
