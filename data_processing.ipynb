```python
import pandas as pd
```

```python
raw = pd.read_csv("raw.csv")
```

```python
raw[['Вторичка', 'Новостройки']] = raw['Вторичка\n\nновостройки\n₽/м²'].str.split('\n', expand=True)
raw[['Аренда 1', 'Аренда 2']] = raw['Аренда 1-комн.\n\n2-комн.\n₽/месяц'].str.split('\n', expand=True)
```

```python
def extract_pattern(text):
    return text.str.extract(r'(^\S+\s+.{3})')[0]
def extract_pattern2(text):
    return text.str.extract(r'(^\S+\s+.{3}\s+.{3})')[0]
```

```python
raw['Аренда 1'] = extract_pattern(raw['Аренда 1'])
raw['Аренда 2'] = extract_pattern(raw['Аренда 2'])
raw['Вторичка'] = extract_pattern(raw['Вторичка'])
raw['Новостройки'] = extract_pattern(raw['Новостройки'])
raw['Квартиры₽/м²'] = extract_pattern(raw['Квартиры₽/м²'])
raw['Земля₽/сотка'] = extract_pattern(raw['Земля₽/сотка'])
```

```python
del raw['Вторичка\n\nновостройки\n₽/м²']
del raw['Аренда 1-комн.\n\n2-комн.\n₽/месяц']
```

```python
raw['Дома₽'] = extract_pattern2(raw['Дома₽'])
```

```python
raw['год'] = 2010 + (raw.index // 12)
raw['Месяц'] = (raw.index % 12) + 1
```
